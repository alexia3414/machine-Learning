```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import*
```

```python
data=pd.read_csv('Loan_application.csv')
```

```python
print(data.head())
```

```python
print(data.isnull().sum())
```

```python
x = data.iloc[:,:-1].values
y=data.iloc[:,-1].values
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
ss=StandardScaler()
x_train[:,[0,1]]=ss.fit_transform(x_train[:,[0,1]])
x_test[:,[0,1]]=ss.transform(x_test[:,[0,1]])
x_train
```

```python
results=LogisticRegression(random_state=0)
results.fit(x_train,y_train)
```

```python
y_pred=results.predict(x_test)
y_pred
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
accuracy_score(y_test,y_pred)
```

```python
recall=recall_score(y_test,y_pred)
recall
```

```python
f1_score(y_test,y_pred)
```

```python
def make_prediction(Age,Salary,Education,Capital):
    y_pred =my_model.predict([[Age, Salary, Education, Capital]])

    if y_pred>0:
        output='congs you are eligible for the loan'
    else:
        output='you are not eligible for the loan'
    return output
```
